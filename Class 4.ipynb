```python
class SinglyLinkedList:
    class Node:
        def __init__(self, data):
            self.data = data
            self.next = None
            
    def __init__(self):
        self.head = None
        
    def append(self, value):
        new_node = self.Node(value)
        if not self.head:
            self.head=new_node
        else:
            current = self.head
            while current.next:
                current = current.next
            current.next = new_node
            
    def __str__(self):
        output = '['
        current = self.head
        if current:
            output += '%s' % current.data
            current = current.next
            while current:
                output += ', %s' % current.data
                current = current.next
        output += ']'
        return output
        
    def index(self, value):
        if self.head:
            counter = 0
            current = self.head
            while current.next and current.value != value:
                current = current.next
                counter += 1
            if current.data == value:
                return counter
        raise ValueError("Value not found")
       
    
    def insert(self ,index, value):
        new_node = self.Node(value)
        if not self.head:
            self.head = new_node
        else:
            if index == 0:
                old_head = self.head
                self.head = new_node
                new_node.next = old_head
            else:
                curent = self.head
                prev = None
                counter = 0
                while current.next and index != counter:
                    prev = current
                    current = current.next
                    counter += 1
                if index == counter:
                    prev.next = new_node
                    new_node.next = current
                else:
                    current.next = new_node
```
